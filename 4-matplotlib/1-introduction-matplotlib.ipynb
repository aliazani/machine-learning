## matplotlib workflow

```python
# A matplotlib workflow

# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig("sample_plot.png")
```

```python
# make a plot(Bar) from dictionary 
food_prices = {'Pizza': 20, 'Egg': 5, 'Pasta': 15}
fig, ax = plt.subplots()
ax.set(title='Foods price', xlabel='Food name', ylabel='Price')
ax.bar(food_prices.keys(), food_prices.values());
```

```python
# crate horizental bar
fig, ax = plt.subplots()
ax.set(title='Food', xlabel='price', ylabel='food-name')
ax.barh(list(food_prices.keys()), list(food_prices.values()));
```

```python
# make some data 
x = np.random.randn(1000)
# create histogram plot
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# subplots ==> 2 options
# 1st:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(figsize=(10, 5), nrows=2, ncols=2)
# plot to each diffrent axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(food_prices.keys(), food_prices.values())
ax4.hist(np.random.randn(10000));
```

```python
# 2nd :
fig, ax = plt.subplots(nrows=2,
                      ncols=2, figsize=(10, 8))
# plot each axis ==> use indexes
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(food_prices.keys(), food_prices.values())
ax[1, 1].hist(np.random.randn(100));
```

## Plotting with pandas using the OO method 

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
over_50.size
```

```python
# pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target', figsize=(10, 10));
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize=(10, 10))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax);
# set the limiter for x
ax.set_xlim([50, 80])
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 10))
# Customize the plot
ax.set(title='Heart disease and Cholestrol levels', xlabel='Age', ylabel='Cholestrol')
# plot the data
scatter = ax.scatter(x=over_50['age'], y=over_50['chol'],
                c=over_50['target'])
# Add legend
ax.legend(*scatter.legend_elements(), title='Target')
# add a horizental line 
ax.axhline(over_50['chol'].mean(), linestyle='-.');
```

## Adding another plot to existing styled one

```python
# sharex = True ==> two plots have same x label so it will be share for both
fig, ax = plt.subplots(figsize=(10, 10), nrows=2, ncols=1, sharex=True)

# ax[0]
scatter = ax[0].scatter(x=over_50['age'], y=over_50['chol'], c=over_50['target'])

ax[0].set(title="Heart Disease and Cholesterol Levels", ylabel='Cholestrol', )
ax[0].legend(*scatter.legend_elements(), title='Target')
ax[0].axhline(y=over_50["chol"].mean(), color='r', linestyle='--', label="Average")
# ax[1]
scatter = ax[1].scatter(x=over_50['age'], y=over_50['thalach'], c=over_50['target'])

ax[1].set(title='Heart disease and Max heart rate', xlabel='Age', ylabel='max heart-rate')
ax[1].legend(*scatter.legend_elements(), title='Target')
ax[1].axhline(y=over_50['thalach'].mean(), color='g', linestyle='-.', label='Average')


# add a title to figure
fig.suptitle('Heart disease Analysis', fontsize=16, fontweight='bold');
```
